```python
from matplotlib import pyplot as plt

# DDA Function for line generation


def DDA(x0, y0, x1, y1):

    # Find absolute differences
    dx = abs(x0 - x1)
    dy = abs(y0 - y1)

    # Find maximum difference
    steps = max(dx, dy)

    # Calculate the increment in x and y
    xinc = dx / steps
    yinc = dy / steps

    # Start with 1st point
    x = float(x0)
    y = float(y0)

    # Make lists for coordinates
    x_coordinates = []
    y_coordinates = []

    for i in range(steps):
        # Append the x,y coordinates in respective lists
        x_coordinates.append(x)
        y_coordinates.append(y)

        # Increment the values
        x += xinc
        y += yinc

    # Plot the line with coordinates list
    plt.plot(x_coordinates, y_coordinates, marker="o", markersize=1, markerfacecolor="green")

    # **This line is missing in the original code**
    plt.show()


# Driver code


# Coordinates of 1st point
x0, y0 = 20, 20

# Coordinates of 2nd point
x1, y1 = 60, 50

# Function call
DDA(x0, y0, x1, y1)

# This code is contributed by 111arpit1
```

4. Develop the program for Bresenhamâ€™s Line drawing algorithm.

```python
import matplotlib.pyplot as plt

def draw_line_bresenham(x1, y1, x2, y2):
    # Calculate differences in coordinates
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    # Determine the sign for x and y
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1

    # Initial decision parameter
    decision_param = 2 * dy - dx

    # Initialize starting coordinates
    x = x1
    y = y1

    # Create lists to store the coordinates of the line
    x_values = []
    y_values = []

    # Perform Bresenham's algorithm
    while x != x2 or y != y2:
        # Add the current point to the line
        x_values.append(x)
        y_values.append(y)

        # Check if decision parameter is positive or negative
        if decision_param > 0:
            decision_param -= 2 * dx
            y += sy

        decision_param += 2 * dy
        x += sx

    # Add the last point to the line
    x_values.append(x)
    y_values.append(y)

    return x_values, y_values

# Example usage:
x1, y1 = 1, 1
x2, y2 = 8, 5

x_values, y_values = draw_line_bresenham(x1, y1, x2, y2)

# Plotting the line
plt.plot(x_values, y_values, marker='o')
plt.title("Bresenham's Line Drawing Algorithm")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

5.Write a program to implement 2D scaling.

```python
import numpy as np
import matplotlib.pyplot as plt


def scale_points(points: np.ndarray, sx: float, sy: float) -> np.ndarray:
    """
    Scale a set of points by sx along the x-axis and sy along the y-axis.
    """
    scaled_points = []
    for point in points:
        scaled_points.append([point[0] * sx, point[1] * sy])
    return np.array(scaled_points)


# Original points
points = np.array([[1, 1], [2, 2], [3, 3]])

# Scaling factors
sx = 2
sy = 0.5

# Scale the points
scaled_points = scale_points(points, sx, sy)

# Plot original and scaled points
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original')
plt.scatter(np.array(scaled_points)[:, 0], np.array(scaled_points)[:, 1], color='red', label='Scaled')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Scaling')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

6. Write a program to perform 2D translation

```python
import numpy as np
import matplotlib.pyplot as plt


def translate_points(points: np.ndarray, dx: float, dy: float) -> np.ndarray:
    """
    Translate a set of points by dx units horizontally and dy units vertically.
    """
    translated_points = []
    for point in points:
        translated_points.append([point[0] + dx, point[1] + dy])
    return np.array(translated_points)


# Original points
points = np.array([[1, 1], [2, 2], [3, 3]])

# Translation amounts
dx = 2
dy = 1

# Translate the points
translated_points = translate_points(points, dx, dy)

# Plot original and translated points
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original')
plt.scatter(np.array(translated_points)[:, 0], np.array(translated_points)[:, 1], color='red', label='Translated')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Translation')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

7.Perform 2D Rotation on a given object

```python
import matplotlib.pyplot as plt
import numpy as np
# Define the vertices of the right-angle triangle
triangle = np.array([[0, 0], [0, 1], [1, 0], [0, 0]])
# Plot the original triangle
plt.plot(triangle[:, 0], triangle[:, 1], label='Original')
# Define the rotation angle (in radians)
angle = np.radians(45)
# Define the rotation matrix
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
# Rotate the triangle
rotated_triangle = np.dot(triangle, rotation_matrix)
# Plot the rotated triangle
plt.plot(rotated_triangle[:, 0], rotated_triangle[:, 1], label='Rotated')
# Set plot labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Rotation of a Right Angle Triangle')
plt.legend()
# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='box')
# Show the plot
plt.grid()
plt.show()
```

```python
import pygame
import random


# Initialize pygame
pygame.init()

# Set screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Text Screen Saver")

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Create a clock object to control the frame rate
clock = pygame.time.Clock()


# Define a function to generate random position, text, and color
def generate_text():
    x = random.randint(0, WIDTH)
    y = random.randint(0, HEIGHT)
    text = random.choice(["Hello, World!", "Welcome to the Screen Saver", "Python is Fun!"])
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    return x, y, text, color


# Main loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Fill the screen with black color
    screen.fill(BLACK)

    # Generate random position, text, and color
    x, y, text, color = generate_text()

    # Create a font object
    font = pygame.font.Font(None, 36)

    # Render the text with the chosen color
    text_surface = font.render(text, True, color)

    # Get the text rectangle
    text_rect = text_surface.get_rect()
    text_rect.center = (x, y)

    # Blit the text onto the screen
    screen.blit(text_surface, text_rect)

    # Update the display
    pygame.display.flip()

    # Control the frame rate
    clock.tick(70)

# Quit pygame
pygame.quit()
```

9. Perform smiling face animation using graphic functions

```python
import pygame
import sys

# Initialize pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 600, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Smiling Face Animation")

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
YELLOW = (255, 255, 0)

# Set up clock
clock = pygame.time.Clock()


# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(BLACK)

    # Draw face
    pygame.draw.circle(screen, YELLOW, (WIDTH // 2, HEIGHT // 2), 100)

    # Draw eyes
    pygame.draw.circle(screen, BLACK, (WIDTH // 2 - 40, HEIGHT // 2 - 30), 15)
    pygame.draw.circle(screen, BLACK, (WIDTH // 2 + 40, HEIGHT // 2 - 30), 15)

    # Draw mouth (smile)
    pygame.draw.arc(screen, BLACK, (WIDTH // 2 - 40, HEIGHT // 2 + 10, 80, 40), 0, 3.14, 2)

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(30)

# Quit pygame
pygame.quit()
sys.exit()
```

10. Draw the moving car on the screen.

```python
import pygame
import sys

# Initialize pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Moving Car")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (128, 128, 128)
RED = (255, 0, 0)

# Car parameters
car_width = 100
car_height = 50
car_x = (WIDTH - car_width) // 2  # Center the car horizontally initially
car_y = HEIGHT - car_height - 20  # Position the car at the bottom
car_speed = 5  # Adjust speed as needed

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Handle key presses
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and car_x > 0:  # Prevent going off-screen (left)
        car_x -= car_speed
    if keys[pygame.K_RIGHT] and car_x < WIDTH - car_width:  # Prevent going off-screen (right)
        car_x += car_speed

    # Clear the screen
    screen.fill(WHITE)

    # Draw the road
    pygame.draw.rect(screen, GRAY, (0, HEIGHT // 2, WIDTH, HEIGHT // 2))

    # Draw the car
    pygame.draw.rect(screen, RED, (car_x, car_y, car_width, car_height))
    pygame.draw.circle(screen, BLACK, (car_x + 20, car_y + car_height), 15)
    pygame.draw.circle(screen, BLACK, (car_x + car_width - 20, car_y + car_height), 15)

    # Update the display
    pygame.display.flip()

# Quit pygame
pygame.quit()
sys.exit()
```
